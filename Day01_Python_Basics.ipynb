```python
# Lists - the most used data structure in data science 
numbers = [10, 25, 3, 47, 8, 92, 14]

print("All numbers:", numbers)
print("First number:", numbers[0])
print("Last number:", numbers[-1])
print("Total count:", len(numbers))
print("Highest:", max(numbers))
print("Lowest:", min(numbers))
print("Sum:", sum(numbers))
print("Average (rounded):", round(sum(numbers) / len(numbers), 2))
```
