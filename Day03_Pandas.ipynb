```python
import pandas as pd
import numpy as np

print("Panda's version:", pd.__version__)
```

```python
data = {
    "Name" : ["Milan", "Rahul", "Bhavdip", "Jayesh", "Vir"],
    "Age"  : [22, 23, 25, 26, 28],
    "City" : ["Ahmedabad", " Rajkot", "Surat", "Jaipur", "Banglore"],
    "Salary" : [50000, 52000, 56000, 54000, 55000],
}

df = pd.DataFrame(data)
print(df)
```

```python
print("Shape:", df.shape)
print("Columns:", df.columns.tolist())
print("Data types: \n", df.dtypes)
print("\nFirst look: \n", df.head())
print("\nBasic stats: \n", df.describe())
```

```python
print("All names:")
print(df["Name"])

print("\nAll salaries:")
print(df["Salary"])

print("\nName and city only:")
print(df[["Name", "City"]])
```

```python
high_salary = df[df["Salary"] > 52000]
print("High earners: \n", high_salary)

ahmedabad = df[df["City"] == "Ahmedabad"]
print("\nFrom Ahmedabad:\n", ahmedabad)

young = df[df["Age"] < 25]
print("\nUnder 25:\n", young)
```

```python
df["Experience"] = [1, 3, 2, 2, 3]
df["Salary_per_year"] = df["Salary"] * 12

print(df)
```

```python
sorted_df = df.sort_values("Salary", ascending=False)
print("Sorted_by_salaries:\n", sorted_df)

print("\nAverage salary by citie:")
print(df.groupby("City")["Salary"].mean())

print("\nPeople per city:")
print(df["City"].value_counts)
```

```python
data2 = {
    "Name" : ["Milan", "Jeet", "Jigo", "Meet", "Migo"],
    "Score" : [99, None, 100, None, 70],
    "Grade" : [None, "B", "A", None, "A"]
    }

df2 = pd.DataFrame(data2)
print("Original with missing values:\n", df2)

print("\nMissing Values:\n", df2.isnull().sum())

df2["Score"] = df2["Score"].fillna(df2["Score"].mean())

df2["Grade"] = df2["Grade"].fillna("A")

print("\nAfter fixing:\n", df2)
```
