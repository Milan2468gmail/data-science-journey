```python
import pandas as pd
import numpy as np

print("Day04 of DataScience!")
```

```python
data = {
    "Name" : ["Milan", "Bhavdip", "Mike", "Ram", "Krishna", "Maruti", "Ganesh"],
    "City" : ["Ahmedabad", "Ahmedabad", "Vegas", "Ayodhya", "Ayodhya", "Anjaney", "Ayodhya"],
    "Department" : ["Data", "Ai", "Data", "Data", "Ai", "Ai", "Ai"],
    "Salary" : [100000,110000,50000,150000,75000,80000,90000],
    "Experience" : [1,2,1,5,4,3,6]

}

df=pd.DataFrame(data)
print(df)
print()

#Average salary by city
print("Average salary by city:")
print(df.groupby("City")["Salary"].mean())
print()

#Count of people by department
print("Count of people by department:")
print(df.groupby("Department")["Name"].count())

#Multiple stats at once
print("Salary stats by department:")
print(df.groupby("Department")["Salary"].agg(["mean", "min", "max"]))
```

```python
employees = pd.DataFrame({
    "EmpID" : [1,2,3,4,5],
    "Name" : ["Milan", "Raj", "Vir", "Kirsh", "Mike"],
    "DeptID" : [101,102,103,101,102]
})

departments = pd.DataFrame({
    "DeptID" : [101,102,103],
    "DeptName" : ["AI team", "Web team", "Data team" ]
})

print("Employees:")
print(employees)
print()
print("Departments:")
print(departments)
print()

#Merge them together like SQL JOIN
merged = pd.merge(employees, departments, on="DeptID")
print("Merged Results are:")
print(merged)
```

```python
df = pd.DataFrame({
    "Name" : ["Milan", "Mike", "Jeet", "Dev", "Raj"],
    "Salary" : [80000,50000,60000,75000,63000],
    "Score" : [98,60,70,80,75]
})

#Apply a custom function to column
def salary_grade(salary) : 
    if salary >= 70000:
        return "High"
    elif salary >= 60000:
        return "Medium"
    else: 
        return "Low"

df["Grade"] = df["Salary"].apply(salary_grade)
print(df)
print()

#Aply with lambda - shorter version
df["Bonus"] = df["Salary"].apply(lambda x: x * 0.10)
print("With Bonus:")
print(df)
```

```python
data = {
    "Name" : ["Milan", "Bhavdip", "Mike", "Ram", "Krishna", "Maruti", "Ganesh"],
    "City" : ["Ahmedabad", "Ahmedabad", "Vegas", "Ayodhya", "Ayodhya", "Anjaney", "Ayodhya"],
    "Department" : ["Data", "Ai", "Data", "Data", "Ai", "Ai", "Ai"],
    "Salary" : [100000,110000,50000,150000,75000,80000,90000],

}

df = pd.DataFrame(data)

#Pivot = Average salary by City and Department
pivot = pd.pivot_table(df, values="Salary", index="City", columns="Department", aggfunc="mean", fill_value=0)
print("Salary Pivot Table:")
print(pivot)
```

```python
data = {
    "City":       ["Ahmedabad", "Mumbai", "Delhi", "Ahmedabad", "Mumbai",
                   "Ahmedabad", "Delhi", "Mumbai", "Ahmedabad", "Delhi"],
    "Score":      [85, 92, 78, 95, 88, 72, 91, 83, 79, 96]
}

df = pd.DataFrame(data)

#Count how many people per city
print("People per City:")
print(df["City"].value_counts())

#Sort by highest to lowest
print("Sortd by score:")
print(df.sort_values("Score", ascending=False))

#Top 3 score only
print("Top 3 scores:")
print(df.sort_values("Score", ascending=False).head(3))
```

```python
data = {
    "Student": ["Milan", "Rahul", "Priya", "Ankit", "Sara",
                "Jeet",  "Migo",  "Meet",  "Rohan", "Neha"],
    "City":    ["Ahmedabad","Mumbai","Delhi","Ahmedabad","Mumbai",
                "Delhi","Ahmedabad","Mumbai","Delhi","Ahmedabad"],
    "Maths":   [85, 92, 78, 95, 88, 72, 91, 83, 79, 96],
    "Science": [90, 85, 92, 88, 76, 95, 82, 89, 93, 77],
    "English": [75, 88, 82, 70, 94, 86, 79, 91, 84, 88]
}

df = pd.DataFrame(data)

#1.Add total and average marks
df["Total"] = df["Maths"] + df["Science"] + df["English"]
df["Average"] = (df["Total"]/3).round(2)
print(df)

#2. Add Gradde based on Average
def get_grade(avg) :
    if avg >= 90 : return "A"
    elif avg >= 80 : return "A"
    elif avg >= 70 : return "B"
    else : return "C"

df["Grade"] = df["Average"].apply(get_grade)

#3. Top Student per city :
print("Top Student per City:")
print(df.groupby("City")["Average"].max())
print()

#4.Class average per subject
print("Class average per subject:")
print(df[["Maths", "Science", "English"]].mean().round(2))
print()

#5. Final sorted results
print("Final Result - Top to Bottom:")
print(df[["Student", "Total", "Average", "Grade"]].sort_values("Total", ascending=False))
print()
```
