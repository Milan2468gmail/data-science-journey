```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#Load all 3 dataset
covid = pd.read_csv("covid_19_india.csv")
vaccine = pd.read_csv("covid_vaccine_statewise.csv")
tested = pd.read_csv("StatewiseTestingDetails.csv")

print("COVID Cases shape:", covid.shape)
print("Vaccine shape:", vaccine.shape)
print("Testing shape:", tested.shape)
```

```python
print("=== COVID CASES ===")
print(covid.head(3))
print("\nColumns:", covid.columns.tolist())

print("\n=== VACCINE DATA ===")
print(vaccine.head(3))
print("\nColumns:", vaccine.columns.tolist())

print("\n=== TESTING DATA ===")
print(tested.head(3))
print("\nColumns:", tested.columns.tolist())
```

```python
print("Info:", covid.info())

print("Info:", vaccine.info())

print("Info:", tested.info())
```

```python
print("\n=== COVID DATA ===")
print(covid.describe(include='all'))

print("\n=== VACCINE DATA ===")
print(vaccine.describe(include='all'))

print("\n=== TESTING DATA ===")
print(tested.describe(include='all'))
```

```python
print("COVID DATA NULL VALUES:\n",covid.isnull().sum())

print("\nVACCINE DATA NULL VALUES:\n",vaccine.isnull().sum())

print("\nTESTING DATA NULL VALUES:\n",tested.isnull().sum())
```

## Which state has the highest confirmed cases?

```python
top_states = covid.groupby("State/UnionTerritory")["Confirmed"].max().reset_index()

top_states = top_states.sort_values(by="Confirmed", ascending=False).head(10)

top_states

```

### Insight 

Maharashtra has the highest number of confirmed COVID_19 cases among all states.
This indicates a higher spread of the virus, possibly due to population density and urbanization.

## Which state has highest deaths?


```python
top_deaths = covid.groupby("State/UnionTerritory")["Deaths"].max().reset_index()

top_deaths = top_deaths.sort_values(by="Deaths", ascending=False).head(10)

top_deaths
```

### Insight

Maharashtra has the highest number of number of COVID_19 deaths among all states.
This is likely driven by its high number of confirmed cases, suggesting a strong correlation between total infection and total deaths.
It may also reflect factors such as populaton density and healthcare system pressure.

## How cases increased over time?

```python
print(covid.groupby("Date")["Confirmed"].sum().plot())
```

### Insight

COVID-19 cases began to rise rapidly from mid-August 2020.
The peak period was observed between March 2021 and June 2021, during wjich the number of confirmed cases increased sognficantly.
This trend likely corresponds to the second wave of COVID-19 in India.

## Which state has highest recovery rate?

```python
covid["Recovery Rate"] = covid["Cured"] / covid["Confirmed"]

print(covid.sort_values("Recovery Rate", ascending=False).head(10))
```

### Insight 

Recovery rates vary across states and are not uniform. 
Ladakh shows one of the highest recovery rates, indicating that a large proportion of confirmed cases have recovered. 
This suggests effective case management relative to its total number of infections.

## Which states have the highest recovery rate?


```python
covid.groupby("State/UnionTerritory")["Recovery Rate"].max().sort_values(ascending=False).head(10)
```

### Insight 

Several states such as Andaman and Nicobar Islands, Goa, Tripura, Mizoram, Kerala, and Ladakh show a recovery rate of 1.0(100%).
This indicates that, at certain points in time, all confirmed cases in these states had recovered.
Such high recovery rates are more common in regions with relatively low total case counts and may not directly reflect overall perfomance compared to larger states.

## Which state has highest death rate?

```python
covid.groupby("State/UnionTerritory")["Death Rate"].max().sort_values(ascending=False).head(10)
```

### Insight

There is significant variation in death rates across states, with Punjab and Bihar showing the highest rates of around 50%. 
However, these values are influenced by very small sample sizes in the early stages of the pandemic. 
As the number of confirmed cases increases, death rates tend to stabilize, making early high percentages less reliable for comparison.

## Final Conclusion

- Maharashtra has the highest number of confirmed COVID-19 cases among all states, indicating a higher spread of the virus.

- COVID-19 cases increased rapidly from mid-August 2020 and reached their peak between March 2021 and June 2021, corresponding to the second wave in India.

- Recovery rates vary across states, with some states showing a 100% recovery rate at certain points in time. However, this is often due to low total case counts rather than overall better performance.

- Death rates also vary across states. Some states show higher death rates, but this is mainly due to small sample sizes during the early stages of the pandemic, where even a single death significantly increases the percentage.

- Overall, the analysis shows that trends in COVID-19 data depend heavily on population size, number of cases, and time period, and should be interpreted carefully.
